```python
df.drop(columns=df.columns[0], axis=1, inplace=True)
df.drop(columns=df.columns[-3], axis=1, inplace=True)
df['CAGR'] = (df['2021']/df['2017'])**(1/5) - 1
df
```

```python
df1 = df.copy()
df1
```

```python
df1['Regular'] = df1['Regular'].map(dict(Yes=1, No=0))
df1['Sugar Free'] = df1['Sugar Free'].map(dict(Yes=1, No=0))
df1['Yellow Edition'] = df1['Yellow Edition'].map(dict(Yes=1, No=0))
df1['Cooler?'] = df1['Cooler?'].map(dict(Yes=1, No=0))
df1['Digital screen?'] = df1['Digital screen?'].map(dict(Yes=1, No=0))
df1['Menu inclusion?'] = df1['Menu inclusion?'].map(dict(Yes=1, No=0))
df1['Posters?'] = df1['Posters?'].map(dict(Yes=1, No=0))
df1
```

```python
cm = sns.light_palette("blue", as_cmap =True)
correlation_analysis = pd.DataFrame(df1[['CAGR' , 'Total Sales', 'Regular',
 'Sugar Free','Yellow Edition', 'Cooler?',
 'Digital screen?','Menu inclusion?',
 'Posters?']]. corr()['CAGR']).reset_index()
correlation_analysis.columns=['Impacting Variable', 'Degree of Linear Impact (Correlation)']
correlation_analysis = correlation_analysis[correlation_analysis['Impacting Variable']!='Amount Collected']
correlation_analysis = correlation_analysis.sort_values('Degree of Linear Impact (Correlation)',ascending=False)
correlation_analysis.style.background_gradient(cmap=cm).set_precision(2)
```

```python
cm = sns.light_palette("green", as_cmap = True)
CA = pd.DataFrame(df1.groupby('Account Type')[['CAGR' , 'Total Sales', 'Regular',
 'Sugar Free','Yellow Edition', 'Cooler?',
 'Digital screen?','Menu inclusion?',
 'Posters?']]. corr()['CAGR']).reset_index()
CA.columns=['Acc Type','Impacting Variable','Correlation']
CA = CA[CA['Impacting Variable']!= 'Amount Collected'].reset_index(drop=True)
CA = CA.sort_values(['Acc Type', 'Correlation'], ascending = False)
CA = CA.dropna()
CA.style.background_gradient(cmap=cm).set_precision(2)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
df1.columns=[mystring.replace(" ", "_") for mystring in df1.columns]
df1.columns=[mystring.replace("?", "") for mystring in df1.columns]
df1.columns=[mystring.replace(")", "") for mystring in df1.columns]
results = smf.ols('CAGR ~ Regular+Sugar_Free+Yellow_Edition+\
 Cooler + Digital_screen + Menu_inclusion+Posters',data=df1).fit()
print(results.summary())
```

```python
df2 = pd.read_html(results.summary().tables[1].as_html(),header=0,index_col=0)[0]
df2 =df2.reset_index()
df2 =df2[df2['P>|t|']<0.05][['index','coef']]
df2
```
